خواسته دو

```python
from sklearn.linear_model import LinearRegression
```

```python
X = company_data['Years'].values.reshape(-1, 1)
Y = company_data['Salary'].values.reshape(-1, 1)
```

```python
model = LinearRegression()
model.fit(X, Y)
```

```python
intercept = model.intercept_[0]
coefficient = model.coef_[0]
```

```python
def predict(X):
    return intercept + coefficient*X
```

```python
predicted_data =  pd.DataFrame({'Years': [i for i in range(2023,2043)],
                               'Salary': [predict(i)[0] for i in range(2023,2043)]})
predicted_data
```

```python
given_data = company_data.iloc[:,0:2]
given_data
```

```python
d = pd.concat([given_data, predicted_data], ignore_index=True)
d
```

خواسته سه

```python
model_predictions = model.predict(X)
model_predictions
```

```python
plt.scatter(company_data['Years'], company_data['Salary'])
plt.plot(company_data['Years'], model_predictions, c='red')
plt.show()
```

```python
score = model.score(X, Y)
score
```
